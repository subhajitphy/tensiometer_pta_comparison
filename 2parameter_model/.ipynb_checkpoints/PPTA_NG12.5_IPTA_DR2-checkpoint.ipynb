# Non-Gaussian tension estimates between PPTA & NG12.5 in IPTADR2 dataset


* The motivation and working principle behind using the tensiometer package for comparing the PTA dataset can be found [here](https://drive.google.com/file/d/1b8BQhM1L_JTt9gAUcCE6jIYZAw3XXBwq/view?usp=share_link).
* The tensiometer metric is developed by Marco Raveri, where he has also provided code along with plenty of documentation that can be applied to any multi-model non-gaussian distribution. To get a detailed flavor of this package please take a look at https://github.com/mraveri/tensiometer .

In this notebook, we show how to compute the statistical significance of a parameter shift between common process posteriors arising from two PTAs [**for e.g: PPTA and NG12.5**] in IPTA DR2 dataset.

We start with two folders for comparison: 1. chains_ppta_dr2, 2. chains_ng12p5. There should be at least two files inside that folder. One of them is the chain and the second one includes the number of parameters that have been considered for the analysis. Next, we need to identify the corresponding common priors to be compared between two PTA datasets. This notebook shows the way to find the tension if the number of common prior between two PTAs is **two**. 
In this notebook, we are comparing only common process HD correlate posteriors between two datasets i.e { logùê¥ ,  ùõæ }, that is described as {gw_gamma gw_log10_A} in the pars.txt file of two folders. We need to pick up the entries of the corresponding chain in the following way.


````python
import sys, os
sys.path.insert(0,os.path.realpath(os.path.join(os.getcwd(),'../..')))
from getdist import plots, MCSamples
from getdist.gaussian_mixtures import GaussianND
import getdist
getdist.chains.print_load_details = False
import scipy
import matplotlib.pyplot as plt
import IPython
from IPython.display import Markdown
import numpy as np
import seaborn as sns
# import the tensiometer tools that we need:
import tensiometer
from tensiometer import utilities
from tensiometer import gaussian_tension
from tensiometer import mcmc_tension
````

### Loading PPTA Chains of IPTA DR2
We are also burning the first 25% of the chain to choose a correct sample of the data since the chain took some time to stabilize for mcmc run. The dataset for PPTA and NG12.5 can be found [here](https://drive.google.com/file/d/1jEKwNIGvc3aic2aiakYom_i1aaQ9mkk3/view?usp=share_link).

````python
ppta_1=np.loadtxt('chains_ppta_dr2/chain_1.txt')
par_ppta=np.loadtxt('chains_ppta_dr2/pars.txt',dtype=str)
len_ppta=len(par_ppta)
ppta_2=ppta_1.T

Ap1=ppta_2[len_ppta-2];gp1=ppta_2[len_ppta-3]

#Burning the chains
Ap=Ap1[int(len(Ap1)/4):];gp=gp1[int(len(Ap1)/4):]
````

### Loading NG12.5 Chains of IPTA DR2

````python
ng_1=np.loadtxt('chains_ng12p5/chain_1.txt')
par_ng=np.loadtxt('chains_ng12p5/pars.txt',dtype=str)
len_ng=len(par_ng)
ng_2=ng_1.T

An1=ng_2[len_ng-1];gn1=ng_2[len_ng-2]
#Burning the chains
An=An1[int(len(An1)/4):];gn=gn1[int(len(An1)/4):]
````

We create an array of names of parameters for comparison as well as the labels of parameters in latex format so that it reflects in the plots.

````python
names = ['logA','gamma']
labels=['\log A','\gamma']
del_names=['delta_logA','delta_gamma']
````

Next, we construct the 2d array of {$\log A$, $\gamma$} from two dataset.

````python
ppta_arr=[Ap,gp];ng_arr=[An,gn]
````

After that, we need to extract the log likelihood from the data so the we can make our array compatible with ``` getdist.mcsamples_ package ```.
Since we burnt first 25% of the data the same need to be applied to Log_likelihood too.

````python
log_L_PPTA=ppta_2[-3][int(len(Ap1)/4):]
log_L_NG12p5=ng_2[-3][int(len(An1)/4):]
````

````python
ppta=MCSamples(samples=np.array(ppta_arr).transpose(),loglikes=log_L_PPTA,names=names,labels=labels,label='PPTA')
ng=MCSamples(samples=np.array(ng_arr).transpose(),loglikes=log_L_NG12p5,names = names,labels=labels,label='NG12.5')
````

#### We plot the 2d posteriors of {$\log A$, $\gamma$} for the two chains

````python
g = plots.get_subplot_plotter()
g.triangle_plot([ppta,ng], filled=True)
plt.savefig('PTAcomparison_IPTADR2.png',dpi=300)
````

Then, We load the package for computing the distribution of the differences between two posteriors, which is given by the convolution integral of two chains. To be more precise,

  * The difference chain (tension) is given by  the convolution integral of the two probability densities ($\color{blue}{\text{Arxiv:1806.04649}}$), $P(\theta_1)$ and $P(\theta_2)$, as

\begin{align}
\label{pdt}
    P(\Delta \theta)=\int_{-\infty}^\infty   \, P_{1} (\theta_1) \, P_{2}(\theta_1-\Delta \theta) \, d \theta_1,
\end{align}
where $\Delta \theta \equiv \theta_1-\theta_2$.  In our case, $\theta \equiv  \{\log A, \gamma\} $.
* The statistical significance of a shift in
parameters is calculated via
\begin{align}
\label{pshift}
    \int_{P(\Delta \theta)>P(0)} P(\Delta \theta) \, \Delta \theta,
\end{align}
where the volume of integration is defined as the region
of parameter space where the probability to get a shift
$\Delta \theta$ is above the isocontour of no shift: $\Delta \theta=0$ ($\color{blue}{\text{Arxiv:1912.04880}}$).



````python
from diff_chain import Diff_chain_shift
````

Now we are ready to plot the distribution of the difference as a contour of the distribution of the difference chains {$1 \sigma$,$2 \sigma$, $\sigma_\mathbf{p}$}, where the shift_probablity ( $\mathbf{p}$ ) is the tension between the two datasets. Note that, the tension between two posteriors is obtained after integrating over the contour corresponding to shift_probability or the isocontour of no shift ($\Delta \theta=0$).

* The "lims" paramter of the function plot_diff_chain is the the x limit and y limit range of the plot in $[x_\text{min},x_\text{max},y_\text{min},y_\text{max}]$ format.

````python
def plot_diff_chain(A1,A2,lims=None,off_size=None):
    diff_chain, nsig, shift_lower, shift_upper=Diff_chain_shift(A1,A2,off_size)
    
    # plot the contour for a sanity check:
    g = plots.get_single_plotter()
    diff_chain.updateSettings({'contours': [shift_probability,0.68, 0.95]})
    g.settings.num_plot_contours = 3
    g.plot_2d(diff_chain,param1=del_names[0], param2=del_names[1], solid_colors='r',filled=True,lims=lims)
    g.add_legend([ A1.getLabel()+' & '+A2.getLabel()+ 
                  '\n\n $\Delta=$'+str("{:.2f}".format(nsig))+'$\sigma$'], colored_text=True);
    g.add_x_marker(0)
    g.add_y_marker(0)
    plt.savefig(A1.getLabel()+'_'+A2.getLabel()+'.png',dpi=300)
    
    return diff_chain,shift_probability
    
    
````

````python
diff_chain,shift_probability=plot_diff_chain(ppta,ng,lims=[-1,2,-4.1,2])
````
